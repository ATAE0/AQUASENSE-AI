```python
from __future__ import absolute_import, division, print_function, unicode_literals
import ydf
import pandas as pd
from google.colab import files
import io
```

La solution de problem en utilisant random forest
J'ai utilisé un bibliothèque que j'ai trouver dans la documentation de tensorflow appeler ydf(Yggdrasil Decision Forests)

```python
uploaded = files.upload()
```

```python
data = pd.read_csv(io.BytesIO(uploaded['aquasense_ai_simulated_data.csv']))
```

```python
training_data = data.iloc[:8000,:]
testing_data = data.iloc[8000:,:]
```

```python
model = ydf.RandomForestLearner(label = 'label', task = ydf.Task.CLASSIFICATION).train(training_data)
```

```python
evaluation = model.evaluate(testing_data)
evaluation
```

```python
print(evaluation.accuracy)
```
